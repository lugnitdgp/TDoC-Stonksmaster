```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
plt.plot(df.RM, df.LSTAT)          #Basic plot which doesn't seem to give any useful insights with default parmaters 
```

```python
plt.plot(df.RM, df.LSTAT, 'ro')    #Basic plot with "o" markers of red colour. This can help us make logical conclusions
```

### The above plot is effectively a scatter plot

```python
plt.scatter(df.RM, df.LSTAT, c = 'b', marker = '+')      #Scatter plot
```

```python
corr = abs(df.corr())
sns.heatmap(corr)
```
