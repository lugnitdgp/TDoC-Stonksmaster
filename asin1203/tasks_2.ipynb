```
import numpy as np
import pandas as pd
```

```
train = pd.read_csv('/content/drive/My Drive/housing.csv')
```

```
train.head(5)
```

LSTAT: Percentage of lower status of the population

PTRATIO: Pupil-teacher ratio by town

RM: Average number of rooms per dwelling

MEDV: Median value of owner-occupied homes in $1000s

Conversion to 2-D array

```
train1 = pd.DataFrame(train['RM']).to_numpy() 
train1
```

```
train2 = pd.DataFrame(train['LSTAT']).to_numpy() 
train2
```

Dropping unnecessary columns

```
train_data=train.drop("PTRATIO",axis=1)
train_data.head(5)
```

Renaming Coloumns

```
train_data.rename(columns={'RM': 'Rooms/dwelling'}, inplace=True)
train_data.sample()
```

```
train_data.rename(columns={'LSTAT': 'Lower-status Pop.(%)'}, inplace=True)
train_data.sample()
```

```
train_data.rename(columns={'MEDV': 'Price(in $1000s)'}, inplace=True)
train_data.sample()
```
