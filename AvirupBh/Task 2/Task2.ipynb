```
import pandas as pd

df=pd.read_csv('housing.csv')
```

```
import matplotlib.pyplot as plt

plt.plot(df['LSTAT'],df['MEDV'],color='blue')
plt.title('MEDV vs LSTAT')
plt.xlabel('LSTAT', fontsize=14)
plt.ylabel('MEDV', fontsize=14)
```

```
plt.plot(df['RM'],df['MEDV'],color='blue')
plt.title('MEDV vs RM')
plt.xlabel('RM', fontsize=14)
plt.ylabel('MEDV', fontsize=14)
```

```
plt.plot(df['PTRATIO'],df['MEDV'],color='blue')
plt.title('MEDV vs PTRATIO')
plt.xlabel('PTRATIO', fontsize=14)
plt.ylabel('MEDV', fontsize=14)
```

```
plt.plot(df['LSTAT'],df['MEDV'],'.')
plt.title('SCATTER PLOT')
plt.xlabel('LSTAT', fontsize=14)
plt.ylabel('MEDV', fontsize=14)
```
