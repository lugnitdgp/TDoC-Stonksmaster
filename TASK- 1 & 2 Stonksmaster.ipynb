## Renaming columns to their full forms

```python
df.rename(columns={"RM": "Avg_num_of_rooms_per_dwelling", "LSTAT": "Lower_status_of_the_population_(percent)", "PTRATIO": "Pupil_to_teacher_ratio","MEDV": "Median_value_of_owner_occupied_homes_in_\$1000s"})
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating Scatter plot of Avg num of rooms per dwelling vs Median value of owner occupied homes in 1000s dollars

```python
fig=plt.figure(figsize=(15,10))
plt.scatter(x=df['RM'],y=df['MEDV'],c ="pink",  
            linewidths = 2,  
            marker ="o",  
            edgecolor ="green",  
            s = 50)
plt.xlabel('Avg_num_of_rooms_per_dwelling')
plt.ylabel('Median_value_of_owner_occupied_homes_in_$1000s')
plt.title('Scatter plot')
```

## Creating Line plot of Pupil to teacher ratio vs Median value of owner occupied homes in 1000s dollars

```python
fig=plt.figure(figsize=(15,10))
plt.plot(df['PTRATIO'],df['MEDV'],c ="red")
plt.xlabel('Pupil_to_teacher_ratio')
plt.ylabel('Median_value_of_owner_occupied_homes_in_$1000s')
plt.title("Line PLot")
```
