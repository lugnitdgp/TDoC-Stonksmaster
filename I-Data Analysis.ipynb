```python
import numpy as np
import pandas as pd
from pandas import read_csv
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"

dataset=read_csv(url)
print(dataset.shape)
print(dataset.head(5))
print(dataset.describe())
plt.hist(dataset['sepal_length'],bins=50,range=[0,20],color='green')
plt.title('sepal-length')
plt.show()
plt.hist(dataset['sepal_width'],bins=100,range=[0,10],color='red')
plt.title('sepal-width')
plt.show()
plt.hist(dataset['petal_length'],bins=600,range=[0,10],color='black')
plt.title('petal-length')
plt.show()
plt.hist(dataset['petal_width'],bins=80,range=[0,5],color='violet',alpha=.9)
plt.title('petal-width')
plt.show()
dataset.plot(kind='box',subplots=True,layout=(2,2),sharex=False,sharey=False)
plt.show()
scatter_matrix(dataset)
plt.show()
```
