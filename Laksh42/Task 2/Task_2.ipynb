```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('housing.csv')
df.head()
```

```python
pd.plotting.scatter_matrix(df)
plt.show()
```

### Scatter plot between MEDV and LSTAT

```python
x= df.LSTAT
y= df.MEDV

plt.scatter(x, y) 
plt.xlabel('LSTAT') 
plt.ylabel('MEDV') 
plt.show()
```

### Line graph between MEDV and LSTAT

```python
plt.plot(x, y) 
plt.xlabel('LSTAT') 
plt.ylabel('MEDV') 
plt.show()
```
