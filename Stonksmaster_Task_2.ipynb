```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df = pd.read_csv("housing.csv")
df.head()
```

```
df.hist(figsize=(15,15))
plt.show()
```

```
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(311)

plt.scatter(np.array(df.iloc[:,0]), np.array(df.iloc[:,3]), s=10, color='red')
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[3])

ax.tick_params(axis='both')

ax = fig.add_subplot(312)

plt.scatter(np.array(df.iloc[:,1]), np.array(df.iloc[:,3]), s=10, color='green')
plt.xlabel(df.columns[1])
plt.ylabel(df.columns[3])

ax.tick_params(axis='both')

ax = fig.add_subplot(313)

plt.scatter(np.array(df.iloc[:,2]), np.array(df.iloc[:,3]), s=10)
plt.xlabel(df.columns[2])
plt.ylabel(df.columns[3])

ax.tick_params(axis='both')


plt.show()
```

```
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(311)

sort_df = df.sort_values(df.columns[0], axis=0)
plt.plot(np.array(sort_df.iloc[:,0]), np.array(sort_df.iloc[:,3]))
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[3])

ax.tick_params(axis='both')

ax = fig.add_subplot(312)

sort_df = df.sort_values(df.columns[1], axis=0)
plt.plot(np.array(sort_df.iloc[:,1]), np.array(sort_df.iloc[:,3]))
plt.xlabel(df.columns[1])
plt.ylabel(df.columns[3])

ax.tick_params(axis='both')

ax = fig.add_subplot(313)

sort_df = df.sort_values(df.columns[2], axis=0)
plt.plot(np.array(sort_df.iloc[:,2]), np.array(sort_df.iloc[:,3]))
plt.xlabel(df.columns[2],)
plt.ylabel(df.columns[3],)

ax.tick_params(axis='both')


plt.show()
```
