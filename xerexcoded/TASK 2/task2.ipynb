```
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```
df=pd.DataFrame(pd.read_csv("housing.csv"))
```

```
df.head()
```

```
x=df['PTRATIO']# sorry for the crazy looking graph , it is evident that this data is not fit for a line graph.
y=df['MEDV']
plt.plot(x,y)
plt.show()
```

```
z=df['LSTAT']# a scatter-plot is more appropopriate for this data .
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(z, x, color='r')
ax.scatter(z, y, color='b')
ax.set_xlabel('PTRATIO')
ax.set_ylabel('MEDV')
ax.set_title('scatter plot')
plt.show()
```
