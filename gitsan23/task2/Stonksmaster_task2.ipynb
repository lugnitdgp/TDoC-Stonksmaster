```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```
df=pd.read_csv("housing.csv")
print(df.head())
```

```
dataset=df.values
dataset
```

```
x=dataset[:,0]
y=dataset[:,-1]
plt.plot(x,y,'.',color='yellow')
plt.show
plt.xlabel('RM')
plt.ylabel('MEDV')
```

```
x=dataset[:,1]
y=dataset[:,-1]
plt.figure(figsize=(15,15))
plt.plot(x,y,'*',color='red')
plt.show
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```
x=dataset[:,2]
y=dataset[:,-1]
plt.figure(figsize=(10,10))
plt.plot(x,y,'.')
plt.show
plt.xlabel('PTRATIO')
plt.ylabel('MEDV')
```
