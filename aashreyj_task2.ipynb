1. Reading CSV file to DataFrame and plotting the scatter matrix to visualize the relationships among different columns of the data, using matplotlib

2. Creating a slice of the original DataFrame which contains records of the 'RM', 'LMSTAT', 'MEDV' columns and then plotting graphs between them using matplotlib library

```
import pandas as pd
from matplotlib import pyplot as plt

# reading the CSV file as a dataframe
df = pd.read_csv('housing.csv')

# creating a custom dataframe as a slice of the original frame
df2 = df.loc[:, ['RM', 'LSTAT', 'MEDV']]
dataset = df2.values

# we have the RM column on the X-axis and the MEDV column on the Y-axis
X = dataset[:, 0]
y = dataset[:, 2]

# plotting the scatter graph
plt.plot(X, y, '.')

# add axis labels
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median Value of Owner-occupied Homes')
plt.title('MEDV v/s RM Graph')

# show the generated graph
plt.show()

# we have the LSTAT column on the X-axis and the MEDV column on the Y-axis
# sorting the X-axis values so that lines are systematic
x = sorted(dataset[:, 1])

# plotting the line graph
plt.plot(x, y)

# add axis labels
plt.xlabel('Lower Status of Population')
plt.ylabel('Median Value of Owner-occupied Homes')
plt.title('MEDV v/s LSTAT')

# show the generated graph
plt.show()
```
